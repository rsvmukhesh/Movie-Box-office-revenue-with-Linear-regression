## Movie Box Office Revenue with linear Regression


###  Basic Linear Function $${h(x)} = \theta_0 + \theta_1{x}$$

```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
data = pandas.read_csv("cost_revenue_clean.csv")
```

```python
data.describe() # Gives us detailed data form csv
```

```python
X = DataFrame(data, columns = ['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
print(X)
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(X, y, alpha = 0.3)
plt.ylim(0, 3e+09)
plt.xlim(0, 45e+07)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worlwide Gross $')
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope coeffcient:

```python
regression.coef_ #slope coefficient tells us for every 1$ we get 3.1 times revenue
```

```python
#Intercept
regression.intercept_
# for movie budget 0$ loss 7M$
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(X, y, alpha = 0.3)
plt.plot(X, regression.predict(X), color="red", linewidth = 4
        ) # relationship btw production budject and movie revenue 
          #as predicted by linear regression model 
plt.ylim(0, 3e+09)
plt.xlim(0, 45e+07)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worlwide Gross $')
plt.show()
```

```python
regression.score(X, y) #amount of variation in films revenue that explains budget
```
